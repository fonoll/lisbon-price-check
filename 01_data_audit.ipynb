```python
import pandas as pd

# загружаем файл в DataFrame (таблицу)
# подставь СВОЁ имя файла — посмотри, как называется csv в папке data/
df = pd.read_csv("data/lisbon-houses.csv")

# ВОПРОС 1: сколько строк?
# shape возвращает (число строк, число колонок)
print("Размер:", df.shape)

# ВОПРОС 2: какие колонки есть?
print("Колонки:", list(df.columns))

# смотрим первые 5 строк глазами — что за данные вообще
df.head()
```

```python
# ВОПРОС 3: сколько пропусков в каждой колонке?
# isna() помечает пустые ячейки как True
# mean() считает долю True = долю пропусков
# *100 переводит в проценты, round(1) округляет
(df.isna().mean() * 100).round(1)
```
